```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\faisa\Downloads\Global_Superstore(CSV).csv", encoding="latin-1")
df.head()
```

```python
df.shape        
df.info()      
df.describe()  
```

```python
df.isnull().sum()          # see which columns have nulls
df["Sales"].fillna(df["Sales"].median(), inplace=True)   # numeric → median
df["City"].fillna("Unknown", inplace=True)               # text → Unknown
df.dropna(subset=["Order ID"])       
```

```python
df.duplicated().sum()      
df.drop_duplicates(inplace=True)
```

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%m/%d/%Y")
df["Ship Date"]  = pd.to_datetime(df["Ship Date"],  format="%m/%d/%Y")
```

```python
print(df[["Order Date", "Ship Date"]].dtypes)
```

```python
df.to_csv("Global_Superstore_Cleaned.csv", index=False)
print("Done! File saved.")
```
